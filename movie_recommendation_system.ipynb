### `ast.literal_eval()` is a Python function used to safely evaluate a string as a Python literal. It provides a secure way to convert strings representing Python data structures (like lists, tuples, dicts, numbers, and booleans) back into their corresponding Python objects.


```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# from cast we will onyl take first 3 names from the dictionary
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
        
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

In crew column we have to select only those dictionary where job=director and then get the name from that dictionary


```python
def fetch_director(obj):
    L=[]

    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
        
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

idia is to convert text into vectors removing stop words i.e text vectorization


```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

CountVectorization reutrns scipy sparse metrix
Converting scipy spars metrix into numpy arrray 


```python

vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# getting the top 5000 common words from the tags
cv.get_feature_names_out()[:1000]
```

```python
# as we can see there are words like action,actions,adventure,adventures basically they r same words 
# lets work on this
# implementing stemming
# Stemming is a process of reducing words to their root form. For example, the words "running," "runs," and "ran" could all be stemmed to the root word "run."
# !pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# now lets again do the vectorization
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
cv.get_feature_names_out()[:1000]
```

```python
# to get the similar movies recommendation we have to calculate the distance between vectors, the vectors which are closer to each other r similar and vice versa
# but here we will not calculate euclidean distance instead we will calculate the angles i.e cosine distance 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
# we get the array of each movie cantaining similarity with all other movies including itself
similarity[121]
```

the idea is to get the index of the movie passsed by user in the dataset
then pass that index to the similarity array 
then sort the similarity distance of that selected index array in decending order but there is a problem,if we sort the similarity array we will lose the similary sequence i.e index position  for that we will use `enumerator` function

The `enumerate()` function in Python returns an iterator that produces tuples of the index and the corresponding value of an iterable object.

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    try:
        movie_index = new_df[new_df['title'] == movie].index[0]
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        for i in movies_list:
            # now lets get name of the movie from the index
            print(new_df.iloc[i[0]].title)
    except IndexError:
        print("Movie not found")
```

```python
recommend('Avatar')
```

```python
recommend('Spectre')
```
